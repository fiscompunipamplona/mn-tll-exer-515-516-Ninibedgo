```
from numpy import linspace
from math import tanh, cosh
from pylab import plot, show, xlabel, ylabel, imshow, hot, xlim, ylim, gray
print('librerias')
```

```
def f(x):                                       # se define la función dependiente x
     return 1 + 0.5 * tanh(2*x)

# calcular df / dx usando el método de diferencia central
def df_dx(x):                                   # se define la función derivada con respecto a x
     h = 10e-5                                  # tamaño de paso
     return (f(x + h) - f(x - h)) / (2*h)

def g(x):                                       # derivada analítica de f (x) anterior
     return 1 / (cosh(2*x) ** 2)

xvalores = linspace(-2, 2, 50)                  # se reaiza el intervalo de x, 50 pasos de -2 a 2 
dfvalores = list(map(df_dx, xvalores))          # valores para dfvalores evaluados en x
gvalores = list(map(g, xvalores))               # valores para gvalores evaluados en x

plot(xvalores, dfvalores, 'o')                  # grafica los resultados, dfvalores y gvalores vs xvalores
plot(xvalores, gvalores)
xlabel('x')
show()                                          # se muestra la grafica
```
